```python
tweets['sentiment'] = tweets['airline_sentiment'].apply(
lambda x: 1 if x == 'negative' else 0)
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
def tweet_to_words(text):
    words = word_tokenize(text.lower())
    stop = set(stopwords.words("english"))
    good_words = [w for w in words if not w in stop]
    return " ".join(good_words)    
```

```python
data = tweets['text'].apply(tweet_to_words)
target = tweets['sentiment']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(data, target)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(ngram_range=(1,2), max_features = 1000)
```

```python
x_train#the numbers in this eg-12603 are the index numbers of the text entries from the table
#these entries are suffled cuztrain_test_split shuffled them
```

```python
train_features = cv.fit_transform(x_train)
type(train_features)
```

```python
cv.get_feature_names()
```

```python
#INTERPRETATION OF THE DOCUMENT-TERM MATRIX
#THIS IS 1 ENTRY OF THIS MATRIX 
#it is basically the frequency of each word of that document(1 entry of x_train) also present in the vocabulary
print(train_features[0])
# the 1st entry in x_train is: @ americanair found footage ! ! : happy .
#word at index 412 in cv.get_feature_names = happy
#similarly at 361 = found, 63 = americanair
```

```python
len(cv.get_feature_names())
```

```python
test_features = cv.transform(x_test)
print(test_features.shape)
print(type(test_features))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier()
clf.fit(train_features, y_train)
```

```python
clf.score(test_features, y_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
clf2 = MultinomialNB()
clf2.fit(train_features, y_train)
clf2.score(test_features, y_test)
```

```python
from sklearn import svm
clf3 =svm.SVC(kernel = "linear")
clf3.fit(train_features, y_train)
print(clf3.score(test_features, y_test))
```
